## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

# Display the data table for preview
combinded_mouse_df = mouse_metadata.merge(study_results, how='left', on='Mouse ID', sort=False)

final_mouse_df = combinded_mouse_df.rename(columns={'Tumor Volume (mm3)': 'Tumor Volume'})
final_mouse_df.head()
```

```python
# Checking the number of mice.
mouse_count = combinded_mouse_df["Mouse ID"].count()
mouse_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_rows = combinded_mouse_df[combinded_mouse_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_rows
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
all_duplicate_rows = combinded_mouse_df[combinded_mouse_df.duplicated('Mouse ID')]
all_duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = combinded_mouse_df.drop_duplicates("Mouse ID")
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mice = clean_df["Mouse ID"].count()
clean_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
mean = combinded_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = combinded_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = combinded_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_dv = combinded_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = combinded_mouse_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
summary_df = pd.DataFrame({"Mean": mean, "Median": median, "Variance": variance, "Standard Deviation": standard_dv, "SEM": sem})
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
Capomulin_df = combinded_mouse_df.loc[combinded_mouse_df["Drug Regimen"] == "Capomulin",:]
Ramicane_df = combinded_mouse_df.loc[combinded_mouse_df["Drug Regimen"] == "Ramicane", :]
Infubinol_df = combinded_mouse_df.loc[combinded_mouse_df["Drug Regimen"] == "Infubinol", :]
Ceftamin_df = combinded_mouse_df.loc[combinded_mouse_df["Drug Regimen"] == "Ceftamin", :]

# Capomulin
Capomulin_last = Capomulin_df.groupby('Mouse ID').max()['Timepoint']
Capomulin_vol = pd.DataFrame(Capomulin_last)
Capomulin_merge = pd.merge(Capomulin_vol, combinded_mouse_df, on=("Mouse ID","Timepoint"),how="left")
Capomulin_merge.head()
```

```python
#Calculate the final tumor volume of each mouse across four of the most promising treatment regimens: 
#Capomulin, Ramicane, Infubinol, and Ceftamin.
#Capomulin calculate the quartiles and IQR and quantitatively determine if there are any potential outliers
Capomulin_tumors = Capomulin_merge["Tumor Volume (mm3)"]

quartiles =Capomulin_tumors.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq


print(f"The lower quartile of Capomulin tumors: {lowerq}")
print(f"The upper quartile of Capomulin tumors: {upperq}")
print(f"The interquartile range of Capomulin tumors: {iqr}")
print(f"The median of Capomulin tumors: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
    
```

```python
Ramicane_last = Ramicane_df.groupby('Mouse ID').max()['Timepoint']
Ramicane_vol = pd.DataFrame(Ramicane_last)
Ramicane_merge = pd.merge(Ramicane_vol, combinded_mouse_df, on=("Mouse ID","Timepoint"),how="left")
Ramicane_merge.head()
```
